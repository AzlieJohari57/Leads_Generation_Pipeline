```python
from apify_client import ApifyClient
import pandas as pd

# Initialize the ApifyClient with your API token
client = ApifyClient("apify_api_0HQ8fc5fw5T1aosdacxKQNQYVBAEwi3tXaJc")

# Prepare the Actor input
run_input = {
    "urls": ["https://www.linkedin.com/company/techhousesg/?originalSubdomain=sg"],
    "sort_order": "relevance",
    "max_employees": 100,
    "proxyConfiguration": {
        "useApifyProxy": True,
        "apifyProxyGroups": ["RESIDENTIAL"],
        "apifyProxyCountry": "US",
    },
}

# Run the Actor and wait for it to finish
run = client.actor("2IMkeDI6D9ZzyLMdX").call(run_input=run_input)

# Fetch Actor results from the run's dataset and save to DataFrame
results = []
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    results.append(item)

# Create DataFrame from results
df = pd.DataFrame(results)
print(f"Scraped {len(df)} records")
df
```
